```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Shruti\Downloads\archive\zomato_dataset.csv")
df.head()
```

```python
df.info()
```

```python
#Cleaning Rating: Text - Float
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```python
#Cleaning Average Delivery Time 
df['Average Delivery Time'] = df['Average Delivery Time'].str.replace(' min', '')
df['Average Delivery Time'] = pd.to_numeric(df['Average Delivery Time'], errors='coerce')
```

```python
#Cleaning Average Price
df['Average Price'] = df['Average Price'].str.replace('₹', '')
df['Average Price'] = df['Average Price'].str.replace(' for one', '')
df['Average Price'] = pd.to_numeric(df['Average Price'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
#Top locations
df['Location'].value_counts().head(10)
```

```python
#Top Cuisines, Fix: Comma - separated
cuisines = df['Cuisine'].str.split(', ')
cuisine_list = cuisines.explode()

cuisine_list.value_counts().head(10)
```

```python
#Average Rating
df['Rating'].mean()
```

```python
#Average Delivery Time
df['Average Delivery Time'].mean()
```

```python
#Average Price 
df['Average Price'].mean()
```

```python
#Rating vs Delivery Time 
df.groupby('Rating')['Average Delivery Time'].mean().sort_index()
```

```python
#Cost vs Rating
df.groupby('Rating')['Average Price'].mean().sort_index()
```

```python
#Top Restaurants
df.sort_values(by='Rating', ascending=False).head(10)
```

```python
#Price Groups
df['Price Category'] = pd.cut(df['Average Price'],
                             bins=[0, 150, 300, 500, 1000],
                             labels=['Low', 'Medium', 'High', 'Premium'])
```

```python
df.groupby('Price Category')['Rating'].mean()
```

```python
#Delivery vs Rating
df.groupby(pd.cut(df['Average Delivery Time'], bins=[0, 20, 30, 40, 60]))['Rating'].mean()
```

```python
#Top cuisines by rating
cuisine_df = df.copy()
cuisine_df['Cuisine'] = cuisine_df['Cuisine'].str.split(', ')
cuisine_df = cuisine_df.explode('Cuisine')

cuisine_df.groupby('Cuisine')['Rating'].mean().sort_values(ascending=False).head(5)
```

```python
#Filtering Cuisines with enough Data
cuisine_counts = cuisine_df['Cuisine'].value_counts()

valid_cuisines = cuisine_counts[cuisine_counts > 50].index

filtered = cuisine_df[cuisine_df['Cuisine'].isin(valid_cuisines)]

filtered.groupby('Cuisine')['Rating'].mean().sort_values(ascending=False).head(5)
```
